```python
import pandas as pd
import time

f="data.csv"
RawData=pd.read_csv(f)
RawData=RawData.drop("Unnamed: 4", axis=1)
RawData=RawData.drop("類別", axis=1)
RawData
```

```python
InterestList=["有興趣",'有意願',"有心動","有機會","有想要","想要"]
InterestDF=None
for interst in InterestList:
    InterestDF=pd.concat([RawData[RawData['客戶事件描述'].str.contains(interst)], InterestDF])
InterestDF
```

```python
UniqeuInterestDF=InterestDF.drop_duplicates(['Unique ID'])
UniqeuInterestDF['Unique ID']
```

獨特的ＩＤ有44位，相較所有的少了25通。

接下來將這些ＩＤ的所有客服抓出來


```python
SpecialClient=list(UniqeuInterstDF['Unique ID'])
SpecialClientCallOut=None
for client in SpecialClient:
    SpecialClientCallOut=pd.concat([RawData[RawData['Unique ID']==client], SpecialClientCallOut])
SpecialClientCallOut
```

```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib
def MakeWordCloud(contents):
    with open('stopwords.txt', 'r') as f:
        stopwords = [line.replace('\n', "") for line in f.readlines()]
    text = []
    for row in contents:
        row = jieba.cut(row)
        for term in row:
            if term not in stopwords:
                text.append(term)
                
    text = ",".join(text)
    
    wc = WordCloud(font_path="./simfang.ttf", #設置字體
               background_color="white",      #背景顏色
               max_words = 100 ,              #文字雲顯示最大詞數
               stopwords=stopwords)           #停用字詞
    wc.generate(text)
    plt.imshow(wc)
    plt.axis("off")
    plt.figure(figsize=(10,10), dpi = 100)
    plt.show()
    
    
```

```python
Callout=list(SpecialClientCallOut['客戶事件描述'])
MakeWordCloud(Callout)
```

```python
content=list(RawData[RawData['客戶事件描述'].str.contains("想要")]['客戶事件描述'])
KeyWord=[]
for each in content:
    index=each.index("想要")
    try:
        pause=each.index("，",index)
    except:
        pause=len(each)
    KeyWord.append(each[index:pause])
KeyWord
```
